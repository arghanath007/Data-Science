```python
import pandas as pd
df= pd.read_csv('/home/arghanath/Documents/Machine Learning/Resources Folder/ML/5_one_hot_encoding/homeprices.csv')
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df, dummies],axis='columns')
merged
```

```python
finalDF= merged.drop(['town','robinsville'], axis='columns')
finalDF
```

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
```

```python
X= finalDF.drop(['price'], axis='columns')
X
```

```python
Y= finalDF.price
Y
```

```python
reg.fit(X,Y)
```

```python
reg.predict([[2800,0,0]])
```

```python
reg.predict([[3400,0,1]])
```

```python
reg.score(X,Y)
```
