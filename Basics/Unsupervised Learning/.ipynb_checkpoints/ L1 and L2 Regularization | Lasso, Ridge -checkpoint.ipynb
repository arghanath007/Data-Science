```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('/home/arghanath/Documents/Machine Learning/Resources Folder/ML/16_regularization/Melbourne_housing_FULL.csv')
dataset
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_needed=['Suburb','Rooms','Type','Method','SellerG', 'Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset= dataset[cols_needed]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset[cols_to_fill_zero]=dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize']=dataset['Landsize'].fillna(dataset['Landsize'].mean())
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset['BuildingArea'].mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset=pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X=dataset.drop('Price', axis=1)
Y= dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train) #Model is overfitted
```
