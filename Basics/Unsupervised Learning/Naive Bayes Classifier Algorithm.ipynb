```python
import pandas as pd
df=pd.read_csv('/home/arghanath/Documents/Machine Learning/Resources Folder/ML/14_naive_bayes/titanic.csv')
df
```

```python
df=df.drop(['PassengerId', 'Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns')
df
```

```python
target=df['Survived']
target
```

```python
inputs=df.drop('Survived', axis='columns')
inputs
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies
```

```python
inputs=pd.concat([inputs, dummies], axis='columns')
inputs
```

```python
inputs=inputs.drop(['Sex'], axis='columns')
```

```python
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs['Age'].head(10)
```

```python
inputs['Age']=inputs['Age'].fillna(inputs['Age'].mean())
inputs['Age'].head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(inputs, target, test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test[:10])
```

```python
model.predict_proba(X_test[:10])
```

```python
dfMail=pd.read_csv('/home/arghanath/Documents/Machine Learning/Resources Folder/ML/14_naive_bayes/spam.csv')
dfMail.head()
```

```python
dfMail.groupby('Category').describe()
```

```python
dfMail['Spam']= dfMail['Category'].apply(lambda x: 1 if x=='spam' else 0)
dfMail
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test=train_test_split(dfMail['Message'], dfMail['Spam'], test_size=0.3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
X_test_count= v.transform(X_test)
model.score(X_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```
