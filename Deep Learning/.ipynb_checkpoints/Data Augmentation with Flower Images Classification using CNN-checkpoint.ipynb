```python
flowers_labels_dict={
    'daisy': 0,
    'dandelion': 1,
    'roses': 2,
    'sunflowers': 3,
    'tulips': 4,
}
```

```python
flowers_images_dict['daisy'][0]
```

```python
img=cv2.imread(str(flowers_images_dict['daisy'][0]))  #OpenCv expects the actual string path, so we have to wrap the WindowsPath within a str().
img
```

```python
img.shape
```

```python
 cv2.resize(img, (180,180)).shape # We need to make all of the images of the same shape before we are training the model, so we have to resize all of the images to a particular shape.
```

```python
X,y= [],[]

for flower_name, images in flowers_images_dict.items():
    for image in images:
        img=cv2.imread(str(image))
        resized_img= cv2.resize(img,(180,180))
        X.append(resized_img)
        y.append(flowers_labels_dict[flower_name])
```

```python
X[:5]
```

```python
y[:5]
```

```python
X=np.array(X)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train_scaled=X_train/255
X_test_scaled=X_test/255
```

```python
X_train_scaled
```

```python
X_test_scaled
```

## Using CNN

```python
num_classes=5 # As we have 5 different types of flowers in the dataset

model=keras.Sequential([
    keras.layers.Conv2D(16,3, padding='same', activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Conv2D(32,3, padding='same', activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Conv2D(64,3, padding='same', activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Flatten(),
    keras.layers.Dense(1000, activation='relu'),
    keras.layers.Dense(num_classes)
])

model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
%%timeit -n1 -r1
with tf.device('/CPU:0'):
    model.fit(X_train_scaled,y_train, epochs=10)
```

```python
%%timeit -n1 -r1
with tf.device('/CPU:0'):
    model.evaluate(X_test_scaled,y_test)
```

### The low score of model.evaluate tells us that the model is overfitted with the training dataset.
