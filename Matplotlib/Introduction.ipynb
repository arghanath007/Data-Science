## Creating Subplots within subplots

#### 1) First way of creating subplots

```python
fig, ((ax0,ax1), (ax2, ax3))= plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Plotting the subplots to each different axis

ax0.plot(x, x/2);
ax1.scatter(np.random.random(20), np.random.random(20));
ax2.hist(X);
ax3.bar(dry_fruits_prices.keys(), dry_fruits_prices.values());
```

 ### 2) Way of creating subplots

```python
fig, ax= plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Plotting the subplots to each different index

ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(20), np.random.random(20));
ax[1,0].hist(np.random.randn(1000));
ax[1,1].bar(dry_fruits_prices.keys(), dry_fruits_prices.values());
```

## Pyplot VS Matplotlib OO method

#### When plotting something quickly, it is okay to use the pyplot method
#### When plotting something more advanced, use the 'OO' method instead.


```python
over_50=heart[heart['age'] >50]
len(over_50)
```

```python
over_50.head()
```

```python
heart.head()
```

#### Pyplot Method

```python
over_50.plot(kind='scatter', x='age', y='chol', c='target')
```

#### OO Method mixed with pyplot method

```python
fix, ax= plt.subplots(figsize=(10,10))
over_50.plot(kind='scatter', x='age',y='chol', c='target', ax=ax)
ax.set_xlim([40,100])
```

#### OO method from scratch

```python
fig, ax= plt.subplots(figsize=(10,10))

# Plotting the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

# Customizing the plot

ax.set(title='Heart Disease and Cholesterol Levels', xlabel='Age', ylabel='Cholesterol');

# Adding a legend to the plot

ax.legend(*scatter.legend_elements(), title='Target');

# Adding a horizontal Line

ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
over_50.head()
```

```python
fig, (ax0,ax1)= plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

# Plotting the ax0 axis

scatter= ax0.scatter(x=over_50['age'],y=over_50['chol'], c=over_50['target'])

#Customize ax0

# ax0.set(title='Heart disease with Cholesterol levels', xlabel='Age', ylabel='Cholesterol')
ax0.set(title='Heart disease with Cholesterol levels', ylabel='Cholesterol')

# Adding a legend and a mean line to the plot

ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), LineStyle='--');

# Doing the plotting of ax1 axis

scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'], c=over_50['target'])

#Customize ax1

ax1.set(title='Heart disease with Thalach levels',xlabel='Age', ylabel='Thalach')

# Adding a legend and a mean line to the plot

ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(), LineStyle='--');

# Add a title to thwe figure

fig.suptitle('Heart Disease Data Analysis', fontsize=16,fontweight='bold');
```

#### Customizing the X and Y axis limitations

```python
fig, (ax0,ax1)= plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

# Plotting the ax0 axis

scatter= ax0.scatter(x=over_50['age'],y=over_50['chol'], c=over_50['target'], cmap='summer')

#Customize ax0

# ax0.set(title='Heart disease with Cholesterol levels', xlabel='Age', ylabel='Cholesterol')
ax0.set(title='Heart disease with Cholesterol levels', ylabel='Cholesterol')

# Changing the X-axis limits

ax0.set_xlim([50,80])

# Adding a legend and a mean line to the plot

ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), LineStyle='--');

# Doing the plotting of ax1 axis

scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'], c=over_50['target'], cmap='winter')

#Customize ax1

ax1.set(title='Heart disease with Thalach levels',xlabel='Age', ylabel='Thalach')

# Adding the X-axis limitation

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Adding a legend and a mean line to the plot

ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(), LineStyle='--');

# Add a title to thwe figure

fig.suptitle('Heart Disease Data Analysis', fontsize=16,fontweight='bold');
```
