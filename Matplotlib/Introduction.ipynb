```python
# 1) import matplotlib and get it ready for plotting in Jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 2) Preparing the data.

X=[1,2,3,4]
Y=[11,22,33,44]

# 3) Setting up the plot(figure)

fig,ax=plt.subplots(figsize=(10,10)) # X,Y values of the graph, weight and height of the graph (weight,height)

# 4) Plotting the data

ax.plot(X,Y)

# 5) Customizing the plotting

ax.set(title='Simple Plot',
xlabel='X-axis', ylabel='Y-axis')

# 6) Saving and Displaying the plot

fig.savefig('./Images/matplotlib_Plot.png')
```

### Bar Graph

```python
dry_fruits_prices={
    'Cashews': 20,
    'Almonds': 25,
    'Walnuts': 50,
    'Rasins': 15,
}

fig,ax= plt.subplots()
ax.bar(dry_fruits_prices.keys(),dry_fruits_prices.values())
ax.set(title="Dry Fruits Prices", ylabel='Prices', xlabel="Dry Fruits");
```

### Horizontal Bar Graph

```python
 dry_fruits_prices={
    'Cashews': 20,
    'Almonds': 25,
    'Walnuts': 50,
    'Rasins': 15,
}

fig,ax= plt.subplots()
ax.barh(list(dry_fruits_prices.keys()),list(dry_fruits_prices.values()))
ax.set(title="Dry Fruits Prices", xlabel='Prices', ylabel="Dry Fruits");
```

### Histogram

## Creating Subplots within subplots

#### 1) First way of creating subplots

```python
fig, ((ax0,ax1), (ax2, ax3))= plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Plotting the subplots to each different axis

ax0.plot(x, x/2);
ax1.scatter(np.random.random(20), np.random.random(20));
ax2.hist(X);
ax3.bar(dry_fruits_prices.keys(), dry_fruits_prices.values());
```

 ### 2) Way of creating subplots

```python
fig, ax= plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Plotting the subplots to each different index

ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(20), np.random.random(20));
ax[1,0].hist(np.random.randn(1000));
ax[1,1].bar(dry_fruits_prices.keys(), dry_fruits_prices.values());
```
