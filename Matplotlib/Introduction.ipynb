```python
X=[1,2,3,4]
Y=[10,133,3453,1231]
```

```python
plt.plot(X,Y)
```

## This way of plotting is called as the stateless way of plotting. It is similar to matlab.

```python
plt.plot(X,Y);
```

### 1st Method

```python
fig=plt.figure()
ax=fig.add_subplot()
plt.show()
```

### 2nd Method

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(X,Y)
plt.show()
```

### 3rd Method (Recommended Method from Docs)

```python
fig, ax=plt.subplots()
ax.plot(X,Y);
```
