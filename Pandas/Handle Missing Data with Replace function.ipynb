```python
import pandas as pd
import numpy as np
df= pd.read_csv('C:\\Users\\JohnWick\\Desktop\\Machine Learning\\Resources Folder\\pandas\\6_handling_missing_data_replace\\weather_data.csv')
df
```

```python
df_new=df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': 0,
}, np.NaN)
df_new
```

```python
df_new1=df.replace({
    '-99999': np.NaN,
    '0': 'Fog'
})
df_new1
```

```python
df_new2=df_new1.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-za-z]'
}, '', regex=True)
df_new2
```
