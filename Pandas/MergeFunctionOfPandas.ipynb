```python
import pandas as pd
df1=pd.DataFrame({
    'city': ['New York', 'Chicago', 'Orlando'],
    'temperature': [21,14,35],
})
df1
```

```python
df2=pd.DataFrame({
     'city': ['Orlando', 'Chicago','New York',],
    'humidity': [68,65,77]
})
df2
```

```python
pd.merge(df1, df2, on='city')
```

```python
df3=pd.DataFrame({
    'city': ['New York', 'Chicago','Orlando', 'Los Angeles'],
    'temperature': [21,14,35, 30],
})
df3
```

```python
df4=pd.DataFrame({
     'city': ['San Fransisco', 'Chicago','New York',],
    'humidity': [60,65,77]
})
df4
```

```python
pd.merge(df3, df4, on='city')
```

```python
pd.merge(df3, df4)
```

```python
pd.merge(df3, df4, on='city', how='outer')
```

```python
pd.merge(df3, df4, on='city', how='left')
```

```python
pd.merge(df3, df4, on='city', how='right')
```

```python
pd.merge(df3, df4, on='city', how='outer', indicator=True)
```

```python
df5=pd.DataFrame({
    'city': ['San Fransisco', 'Chicago','New York',],
    'temperature': [21,14,35],
    'humidity': [40,55,27],
})
df5
```

```python
df6=pd.DataFrame({
    'city': ['San Fransisco', 'Chicago','New York',],
    'temperature': [25,19,30],
    'humidity': [50,45,47],
})
df6
```

```python
df7=pd.merge(df5,df6,on='city', suffixes=['_left', '_right'])
df7
```
